```python
from langgraph.graph import StateGraph, START,END
from typing import TypedDict
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
llm_model = ChatGoogleGenerativeAI(
    model = "gemini-2.5-flash",
    temperature = 0.7
)
```

```python
# 1. define state

class Blogtate(TypedDict):
    topic: str
    outline: str
    content: str
    
```

```python
def write_outline(state: Blogtate) -> Blogtate:
    topic = state['topic']
    
    # call llm to generae outline
    prompt = f"write an outline for a blog post about {topic}"
    outline = llm_model.invoke(prompt).content
    
    state['outline'] = outline
    
    return state 
```

```python
# function for blog_content
def write_content(state: Blogtate) -> Blogtate:
    topic = state['topic']
    outline = state['outline']
    
    # call llm to generate content
    prompt = f"write a blog post for the topic {topic} based on the outline:\n {outline}"
    content = llm_model.invoke(prompt).content
    
    state['content'] = content
    
    return state
```

```python
# define graph

graph = StateGraph(Blogtate)
graph.add_node('write_outline',write_outline)
graph.add_node('write_content',write_content)

graph.add_edge(START,'write_outline')
graph.add_edge('write_outline','write_content')
graph.add_edge('write_content',END)

workflow = graph.compile()
```

```python
initial_state = {
    "topic": "Machine Learning"
}

final_state = workflow.invoke(initial_state)
print(final_state)
```
